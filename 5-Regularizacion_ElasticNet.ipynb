#### **3. Técnica ElasticNet**

## Usaremos ElasticNet cuando tengamos un gran número de atributos. Algunos de ellos serán irrelevantes y otros estarán correlados entre ellos.

**Para aplicar este método hay que estandarizar SI O SI.**

## **Antes de aplicar la técnica ElasticNet haremos una regresión lineal de nuestros datos CON ESTANDARIZACIÓN INCLUÍDA.** Esto quiere decir que cogeremos X_train e y_train de la regresión lineal previa en la que ya los hemos estandarizado. Nada de crear nuevos X_train o y_train

**1. Importamos el objeto Lasso**

```python
from sklearn.linear_model import ElasticNetCV
```

**2. Asigna el objeto Lasso a una variable con varias alphas, ya decidirá él cual es la que más le conviene**

```python
e_net = ElasticNetCV(l1_ratio=[0, 0.1, 0.5, 0.7, 0.9, 0.95, 0.99],alphas=np.logspace(-10, 3, 200))
```

**3. Entrena el objeto ElasticNet**

```python
e_net.fit(x_train, y_train)
```

**4. Mira a ver como han quedado los coeficientes para la mejor alpha**

```python
print("Coeficientes:", e_net.coef_)
print("Interceptor:", e_net.intercept_)
```

 - Puedes ver cual ha sido la mejor alpha y el mejor l1

```python
print(f"Mejor valor de alpha encontrado: {e_net.alpha_}")
print(f"Mejor valor de l1_ratio encontrado: {e_net.l1_ratio_}")
```

```python
df = pd.read_csv(r'ruta de tu csv')

X = df['ColumnaS (suele ser en plural) que son las features'] --> Define las variables que tienes pa predecir

y = df['La columna (solo una, campeón/a) que es el target'] --> Define el objetivo que tienes pa predecir
```

## He sido un pesao con la estandarizacion. Aquí ya no hace falta. Van los datos a pelo shempion.

1. Venga, importa de la librería sklearn el paquete de separar los datos, que ahorra un curro que no veas.

```python
from sklearn.model_selection import train_test_split
```

2. A separar datos (joer, esto al final es siempre lo mismo)

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

3. **Empezamos ya a hacer la regresión logística como tal, para eso primero importala Hulio.**

```python
from sklearn.linear_model import LogisticRegression --> Importa el objeto regresión logística
```

4. Crea el objeto de regresion y entrénalo

```python
logisticRegr = LogisticRegression() --> Crea el objeto

logisticRegr.fit(X_train, y_train) --> Entrena la regresion logística
```

5. Predice lo que quieras (por lo general suele ser x_test)

```python
logisticRegr.predict(X_test)
```

6. Si quieres puedes sacar las probabilidades de la predicción

```python
logisticRegr.predict_proba(X_test)
```

7. **Mide el desempeño del modelo**

```python
score = logisticRegr.score(X_test, y_test)
print(score * 100, "%")
```

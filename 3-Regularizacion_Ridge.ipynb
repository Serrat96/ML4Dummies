# **Regularización**

## ¿Qué es la regularización?

### La regularización se aplica básicamente para evitar la sobreoptimización de modelos. **Para aplicar estos métodos hay que estandarizar SI O SI.** Hay 3 técnicas:

#### **1. Técnica Ridge**

Tiene como objetivo el reducir todos los coeficientes del modelo sin que ninguno de ellos llegue a cero. Esto **es útil para reducir mucho la sobreoptimización (overfitting) sin aumentar la infraoptimización (underfitting).** 

Con este método, el coeficiente (la importancia) de nuestras variables **nunca llegará a cero,** lo que quiere decir que **ninguna variable será eliminada de la predicción.** Tendrán un coeficiente más alto (serán más importantes para predecir) o tendrán un coeficiente más bajo (serán menos importantes para predecir) pero su importancia nunca será 0.

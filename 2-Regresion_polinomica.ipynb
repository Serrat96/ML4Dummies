## **2.Regresión polinómica**

**Antes de hacer la regresión polinómica se tiene que hacer una regresión lineal y ver que no es apta para lo que queremos predecir, por ello no se van a crear las X ni las y en este notebook. Ya se crean en el anterior notebook de regresión lineal.**

Una regresión polinómica **trata de añadirle complejidad a la regresión lineal para poder ajustarte mejor a tu modelo cuando la regresión lineal es insuficiente.**

Para empezar, importamos el objeto "PolunomialFeatures" para poder hacer la regresión polinómica

```python
from sklearn.preprocessing import PolynomialFeatures
```

**Creamos el objeto de regresión polinómica** con el número de grados que queremos que haga. **Con más de 4 grados tenemos una posibilidad muy alta de caer en sobreoptimización** (overfitting pa los flipaos)

```python
poly_reg = PolynomialFeatures(degree=3)
```

Entrenamos nuestra regresión polinómica

```python
poly_reg.fit(X)
```

Aplicamos la transformación que nos permitirá graficar todo más adelante como si fuese una regresión lineal normal

```python
X_poly = poly_reg.transform(X)
```

# **Árboles de decisión**

## ¿Para qué sirve esto?

**Un árbol de decisión básicamente es una serie de preguntas que haces a los datos a las que vas respondiendo si o no. En función de lo que vayas respondiendo el dato se irá clasificando en un grupo u otro**

**Dentro de los árboles de decisión, hay dos subgrupos:**

- Árboles de regresión (notebook 7.1)
- Árboles de clasificación (notebook 7.2)
